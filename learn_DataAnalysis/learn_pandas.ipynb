```python
# 数据分组，分组运算
```

```python
group = df.groupby(df['产地'])
group
```

```python
group.mean()
```

```python
# 计算每年的平均评分
df['评分'].groupby(df['年代']).mean()
```

```python
# 只会对数值变量进行分组运算
df['年代'] = df['年代'].astype('str') # 将年份转成字符串就不会像上面对年份进行均值的运算
```

```python
df.groupby(df['产地']).mean()
```

```python
# 传入多个分组变量
df.groupby([df['产地'], df['年代']]).mean()
```

```python
# 获取每个地区，每一年的电影评分的均值
group = df['评分'].groupby([df['产地'], df['年代']])
means = group.mean()
means
```

```python
# 离散化处理，离散化也称分组、区间化
# 在实际的数据分析项目中，对所有的数据属性，我们往往不关注数据的绝对取值，只关注它所处的区间或者等级。
```

```python
df['评分等级'] = pd.cut(df['评分'], [0,3,5,7,9,10], labels=['E','D','C','B','A'])
df
```

```python
# 同样的投票人数也可以刻画电影的热门程度
bins = np.percentile(df['投票人数'], [0,20,40,60,80,100])
```

```python
df['热门程度'] = pd.cut(df['投票人数'], bins, labels=['E','D','C','B','A'])
df
```

```python
df[(df['热门程度']=='A')&(df['评分等级']=='E')]
```

```python
df[(df['热门程度']=='E')&(df['评分等级']=='A')]
```
