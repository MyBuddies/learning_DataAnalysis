## 1.MatPlotLib基础

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4])
plt.ylabel('y')
plt.xlabel('x')
```

```python
# 指定x,y
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'g*')
plt.show()
```

```python
# 调整显示范围
plt.plot([1,2,3,4],[1,4,9,16])
plt.axis([0,6,0,20])
plt.show()
```

```python
# 传入numpy数组
t = np.arange(0.,5.,0.2)
plt.plot(t,t,'r--',
         t,t**2,'bs',
         t,t**3,'g^')
plt.show()
```

```python
# 线条属性
x = np.linspace(-np.pi, np.pi)
y = np.sin(x)
plt.plot(x, y, linewidth=4.0, color='r')
plt.show()
```

```python
# 使用plt.plot()的返回值来设置线条属性
# plot函数返回一个Line2D对象组成的列表，每个对象代表输入的一对组合
```

```python
line1, line2 = plt.plot(x,y,'r-',
                        x,y+1,'g-')
line1.set_antialiased(False)
plt.show()
```

```python
# plt.setp()修改线条性质
line = plt.plot(x, y)
plt.setp(line, color='g', linewidth=4)
```

```python
# 子图
# figure()会产生一个指定编号为num的图
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(figsize=(10, 6))
plt.subplot(211)
plt.plot(t1, f(t1), 'bo',
         t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

## 2.电影数据绘图

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 改变字体乱码
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
df = pd.read_excel('./movie_data3.xlsx')
```

```python
df
```

```python
# (1)绘制每个国家或地区的电影数量的柱状图
```

```python
data = df['产地'].value_counts()
data
```

```python
x = data.index
y = data.values

plt.figure(figsize=(10,6))
plt.bar(x, y, color='g')

plt.title('每个国家或地区的电影数量', fontsize=20)
plt.xlabel('国家或地区', fontsize=18)
plt.ylabel('电影数量', fontsize=18)
plt.tick_params(labelsize=14)
plt.xticks(rotation=90)

for a, b in zip(x,y):
    plt.text(a, b+10, b, ha='center', va='bottom', fontsize=10)


plt.grid() # 网格线

plt.show()
```

```python
# (2)绘制每年上映的电影数量的曲线图
```

```python
# 1888-2015年
data = df['年代'].value_counts()
data = data.sort_index()[:-2]
data
```

```python
x = data.index
y = data.values

plt.plot(x, y, color='b')
plt.title('每年电影数量', fontsize=20)
plt.xlabel('年份', fontsize=18)
plt.ylabel('电影数量', fontsize=18)

for a, b in zip(x[::10], y[::10]):
    plt.text(a, b+10, b, ha='center', va='bottom', fontsize=10)

plt.annotate('2012年达到最大值', xy=(2012,data[2012]), xytext=(2025,2100), arrowprops=dict(facecolor='black', edgecolor='red'))

plt.text(1980,1000,'电影数量开始快速增长')

plt.show()
```

```python
# (3)根据电影的长度绘制饼图--适用分类比较少的
```

```python
data = pd.cut(df['时长'], [0,60,90,110,1000]).value_counts()
data
```

```python
y = data.values
y = y/sum(y) # 归一化

plt.figure(figsize=(7,7))
plt.title('电影时长占比', fontsize=15)
patches, l_text, p_test = plt.pie(y, labels=data.index, autopct='%.1f %%', colors='bygr', startangle=90)

for i in p_test:
    i.set_size(15)
    i.set_color('w')

for i in l_text:
    i.set_size(15)
    i.set_color('r')

plt.legend()

plt.show()
```

```python
# (4)根据电影的评分绘制频率分布直方图
```

```python
plt.figure(figsize=(10,6))

plt.hist(df['评分'], bins=20, edgecolor='k', alpha=0.5)

plt.show()
```

```python
# 双轴图
```

```python
import matplotlib.mlab as mlab
from scipy.stats import norm
```

```python
fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(111)
n, bins, patches = ax1.hist(df['评分'], bins=100, color='m')

ax1.set_ylabel('电影数量', fontsize=15)
ax1.set_xlabel('评分', fontsize=15)
ax1.set_title('频率分布图', fontsize=20)

# 曲线拟合
y = norm.pdf(bins, df['评分'].mean(), df['评分'].std())

# 双轴图
ax2 = ax1.twinx()
ax2.plot(bins, y, 'b--')
ax2.set_ylabel('概率分布', fontsize=15)

plt.show()
```

```python
# (5)根据电影时长和电影评分绘制散点图
```

```python
x = df['时长'][::100]
y = df['评分'][::100]

plt.figure(figsize=(10,6))
plt.scatter(x, y, color='c', marker='p')
plt.legend()
plt.title('电影时长和电影评分散点图', fontsize=20)
plt.xlabel('时长', fontsize=18)
plt.ylabel('评分', fontsize=18)
plt.show()
```

```python
# (6)绘制各个地区的评分箱型图
```

```python
# 美国电影评分箱线图
data = df[df['产地']=='美国']['评分']

plt.figure(figsize=(10,6))
plt.boxplot(data, whis=2, flierprops={'marker':'o','markerfacecolor':'r','color':'k'},
            patch_artist=True, boxprops={'color':'k','facecolor':'#9999ff'})
plt.title('美国电影评分', fontsize=20)

plt.show()
```

```python
# 多组数据的箱线图
data1 = df[df['产地']=='中国大陆']['评分']
data2 = df[df['产地']=='日本']['评分']
data3 = df[df['产地']=='中国香港']['评分']
data4 = df[df['产地']=='英国']['评分']
data5 = df[df['产地']=='法国']['评分']

plt.figure(figsize=(10,8))
plt.boxplot([data,data1,data2,data3,data4,data5],
            labels=['美国','中国大陆','日本','中国香港','英国','法国'],
            whis=2, 
            flierprops={'marker':'o','markerfacecolor':'r','color':'k'},
            patch_artist=True, 
            boxprops={'color':'k','facecolor':'#9999ff'})

ax = plt.gca()
ax.patch.set_facecolor('gray')
ax.patch.set_alpha(0.3)

plt.title('电影评分箱线图', fontsize=20)

plt.show()
```

```python
# 通过vert将图旋转过来
```

```python
# 多组数据的箱线图
data1 = df[df['产地']=='中国大陆']['评分']
data2 = df[df['产地']=='日本']['评分']
data3 = df[df['产地']=='中国香港']['评分']
data4 = df[df['产地']=='英国']['评分']
data5 = df[df['产地']=='法国']['评分']

plt.figure(figsize=(10,8))
plt.boxplot([data,data1,data2,data3,data4,data5],
            labels=['美国','中国大陆','日本','中国香港','英国','法国'],
            whis=2, 
            flierprops={'marker':'o','markerfacecolor':'r','color':'k'},
            patch_artist=True, 
            boxprops={'color':'k','facecolor':'#9999ff'},
            vert=False)# 通过vert将图旋转过来

ax = plt.gca()
ax.patch.set_facecolor('gray')
ax.patch.set_alpha(0.3)

plt.title('电影评分箱线图', fontsize=20)

plt.show()
```

```python
# 画电影时长，投票人数，评分的一个相关系数矩阵图
```

```python
import seaborn as sns

corr = data.corr()
corr = abs(corr)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)

ax = sns.heatmap(corr, vmax=1, vmin=0, annot=True, annot_kws={'size':13,'weight':'bold'}, linewidths=0.05)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.show()
```
